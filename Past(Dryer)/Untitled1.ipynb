```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back - 1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i+look_back, 0])
    return np.array(dataX), np.array(dataY)

np.random.seed(7)
```

```python
dataframe = pd.read_csv('./data/train_keras.csv', engine='python')
```

```python
df_all_np = dataframe.values.astype(np.float)
scaler = MinMaxScaler(feature_range=(0, 1))

device = df_all_np[:, 1:2]
norm_device = scaler.fit_transform(device)

temp = df_all_np[:, 2:3]
norm_temp = scaler.fit_transform(temp)

hum = df_all_np[:, 3:4]
norm_hum = scaler.fit_transform(hum)

day = df_all_np[:, -1:]
norm_day = scaler.fit_transform(day)

dataset = np.concatenate((norm_temp, norm_hum, norm_day, norm_device), axis=1)
```
